```python
def add_lat_lon(path: str, comb_df: pd.DataFrame) -> pd.DataFrame:
    """Add latitude and longitude to combined demographic dataframe

    Args:
        path (str): location of lat lon csv file
        comb_df (pd.DataFrame): combined demographic datafrmae

    Returns:
        pd.DataFrame: combined demographic dataframe with latitude and longitude

    """
    df = pd.read_csv(path)
    return comb_df.merge(df, on='zip', how='left')
newdf = add_lat_lon('zip_lat_lon.csv', df)
```

```python
import statsmodels.formula.api as smf
df2 = newdf[['percent_asian', 'count_asian', 'pop_total', 'lat', 'lon', 'url_length' ]]
df2 = df2.dropna()
df2.head()
def standardize(ser):
  mx = ser.max()
  mn = ser.min()
  return (ser - mn)/(mx - mn)
df2['percent_asian'] = standardize(df2['percent_asian'])
df2['pop_total'] = standardize(df2['pop_total'])
df2['lat'] = standardize(df2['lat'])
df2['lon'] = standardize(df2['lon'])
df2['url_length'] = standardize(df2['url_length'])

md = smf.poisson('count_asian ~ percent_asian + pop_total + lat + lon + url_length', data=df2) #poisson
md = md.fit()
print(md.summary())
```

```python
import statsmodels.formula.api as smf
md = smf.ols('count_asian ~ percent_asian + pop_total +lat + lon', data=df2) #ordinary least squares
md = md.fit()
print(md.summary())
```

```python
scorecard_column_names = {
    'INSTNM': 'university',
    'ENDOWBEGIN': 'endowment',
    'UGDS_ASIAN': 'percent_college_asian',
}


def add_endowment(path: str, comb_df: pd.DataFrame) -> pd.DataFrame:
    """Add in college endowments from scorecard

    Args:
        path (str): location of college scorecard
        comb_df (pd.DataFrame): combined dataframe containing previous data

    Returns:
        pd.DataFrame: output with endowment added
    """
    df = pd.read_csv(path)
    df = df[[key for key in scorecard_column_names]].rename(columns = scorecard_column_names)
    df['university'] = df['university'].str.lower()
    comb_df['university'] = comb_df['university'].str.lower()
    return comb_df.merge(df, how='left', on='university')
newdf = add_endowment('Most-Recent-Cohorts-Institution.csv', newdf)
```
